```python
nan_columns = df.columns[df.isna().any()].tolist()

for col in nan_columns:
    if df[col].dtype == 'object':
        for i, val in enumerate(df[col].dropna()):
            if any(ord(char) < 32 for char in str(val)):
                print(f"⚠️ Control character in column {col}, row {i}: {repr(val)}")
# Finding control character causing error in visualization
```

```python
def remove_control_chars(s):
    return ''.join(c for c in str(s) if ord(c) >= 32)

df['ERC20 most sent token type'] = df['ERC20 most sent token type'].apply(remove_control_chars)
df['ERC20_most_rec_token_type'] = df['ERC20_most_rec_token_type'].apply(remove_control_chars)
```

```python
# Recheck
nan_columns = df.columns[df.isna().any()].tolist()

for col in nan_columns:
    if df[col].dtype == 'object':
        for i, val in enumerate(df[col].dropna()):
            if any(ord(char) < 32 for char in str(val)):
                print(f"⚠️ Control character in column {col}, row {i}: {repr(val)}")
# Finding control character causing error in visualization
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import warnings

warnings.filterwarnings("ignore", category=FutureWarning)

for col in nan_columns:
    plt.figure(figsize=(6, 4))
    sns.histplot(df[col], kde=True, bins=50)
    plt.title(f'Distribution of {col}')
    plt.xlabel(col)
    plt.ylabel("Frequency")
    plt.tight_layout()
    plt.show()
```

```python
# Group all the columns which have the median ~ mean
import numpy as np

median_fill = []
tol = 1e-6  # epsilon

for col in nan_columns:
    if pd.api.types.is_numeric_dtype(df[col]):
        mean = df[col].mean(skipna=True)
        median = df[col].median(skipna=True)
        if np.isclose(mean, median, atol=tol):
            median_fill.append(col)

print("Columns which have mean ≈ median:")
print(median_fill)
```

```python
len(median_fill)
```

```python
# Recheck if these columns have mean ~ median, visualize it
warnings.filterwarnings("ignore", category=FutureWarning)

for col in median_fill:
    plt.figure(figsize=(6, 4))
    sns.histplot(df[col], kde=True, bins=50)
    plt.title(f'Distribution of {col}')
    plt.xlabel(col)
    plt.ylabel("Frequency")
    plt.tight_layout()
    plt.show()
```

```python
# Fill nan these columns with median
for col in median_fill:
    df[col].fillna(df[col].median(), inplace=True)
```

## Finding numeric columns and cat columns for encoding

```python
numeric_cols = df.select_dtypes(include='number').columns.tolist()
cat_cols = df.select_dtypes(include='object').columns.tolist()
```

```python
print(len(numeric_cols))
print(len(cat_cols))
```

```python
cat_cols
```

```python
# Drop address because entropy is too high, it's unique value for each transaction
df = df.drop(columns=['Address'])
```

```python
cat_cols.remove('Address')
```

```python
print(cat_cols)
```

```python
for col in cat_cols:
    unique_value = df[col].nunique()
    print(f'{col} has {unique_value} unique values')
```

```python
encoders = {}
```

```python
# With large unique values => Label Encoding/Count Encoding/Target Encoding
# Let's try with Label Encoding first
from sklearn.preprocessing import LabelEncoder

for col in cat_cols: 
    print(f"-> Đang xử lý cột: {col}")
    # Tạo một encoder mới cho cột này
    le = LabelEncoder()
    
    # Fit và transform cột
    df[col + '_label'] = le.fit_transform(df[col].astype(str))
    
    # Lưu encoder đã được 'fit' vào dictionary
    # Key là tên cột gốc, value là đối tượng encoder
    encoders[col] = le
```

```python
# Drop columns after encoding
df = df.drop(columns=cat_cols)
```
